```python
import requests

download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"

# สร้างชื่อไฟล์ csv
target_csv_path = "nba_all_elo.csv"

response = requests.get(download_url)
response.raise_for_status()    # Check that the request was successful
# เปิดไฟล์
with open(target_csv_path, "wb") as f:
    f.write(response.content)
print("Download ready.")
```

```python
import pandas as pd

data = pd.read_csv("nba_all_elo.csv")

df = pd.DataFrame(data)

display(df)
# การใช้ .to_datetime
# เข้าถึงคอลัมน์ "date_game"
df['date_played'] = pd.to_datetime(df["date_game"])
# display(df['date_played'])

# เข้าถึงโดยใช้ .loc[ชื่อตัวแปรเก็บข้อมูล[ชื่อคอลัมน์]] 
print("เข้าถึงโดยใช้ .loc[]")
display(df.loc[df["team_id"] == "MNL" , "date_played"].min())
display(df.loc[df["team_id"] == "MNL" , "date_played"].max())

# aggregate functions
display(df.loc[df["team_id"] == "MNL", "date_played"].agg(("min", "max")))

display(df.loc[df["team_id"] == "BOS", "pts"].sum())
```
